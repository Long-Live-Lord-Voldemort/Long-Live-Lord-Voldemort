```python
import scipy.stats as sct
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
import seaborn as sb
from matplotlib import rcParams
```

**Discounted Cash Flow or Dividend Discount Model(in Perpetuity or forever)**

**Gordon Growth Model**

```python
d0 = 100
df =[]

for i in range(5000):
  g = random.uniform(0.02, 0.08)
  r = random.uniform(0.09, 0.12)
  d1 = d0*(1+g)
  v0 = d1/(r - g)
  df.append(v0)

rcParams['figure.figsize'] = 15,10
sb.distplot(df)
plt.ylabel('Frequency')
plt.xlabel('Market Value')
```

```python
d0 = 100
g_mean = 0.05
g_sd = 0.01

df = []

for i in range(5000):
  g = sct.norm.ppf(random.uniform(0,1), loc=g_mean, scale=g_sd)
  r = 0.10
  d1 = d0*(1 + g)
  v0 = d1/(r - g)
  df.append(v0)
```

```python
rcParams['figure.figsize'] = 15,10
sb.distplot(df)
plt.ylabel('Frequency')
plt.xlabel('Market Value')
```
