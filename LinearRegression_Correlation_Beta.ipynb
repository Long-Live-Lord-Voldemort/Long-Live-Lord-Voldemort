```python
ticks = ["AAPL", "MSFT", "TSLA", "IBM", "SPY"]
dt = yf.download(ticks, start="2015-01-01", end="2022-01-05", rounding= True)["Adj Close"]
```

```python
dt.tail(11)
```

```python
log_returns = np.log(dt/dt.shift())
```

```python
log_returns.tail(11)
```

```python
import numpy as np
```

```python
tick_a = 'AAPL'
tick_b = 'SPY'



X = log_returns[tick_a].iloc[1:].to_numpy().reshape(-1,1)
Y = log_returns[tick_b].iloc[1:].to_numpy().reshape(-1,1)

lin_regr = LinearRegression()
lin_regr.fit(X,Y)

Y_pred = lin_regr.predict(X)


alpha = lin_regr.intercept_[0]
beta = lin_regr.coef_[0,0]


fig, ax = plt.subplots()
ax.set_title("alpha" + str(round(alpha)))
ax.scatter(X,Y, alpha=0.25)
ax.plot(X,Y_pred, c='green', alpha=0.6)
```

```python
def LinearRegression(tick_a, tick_b):
  X = log_returns[tick_a].iloc[1:].to_numpy().reshape(-1,1)
  Y = log_returns[tick_b].iloc[1:].to_numpy().reshape(-1,1)

  lin_regr = LinearRegression()
  lin_regr.fit(X,Y)

  Y_pred = lin_regr.predict(X)


  alpha = lin_regr.intercept_[0]
  beta = lin_regr.coef_[0,0]


  fig, ax = plt.subplots()
  ax.set_title("alpha" + str(round(alpha, 5)) + "beta:" + str(round(beta, 3)))
  ax.scatter(X,Y, alpha=0.25)
  ax.plot(X,Y_pred, c='green', alpha=0.6)
```
